```python
audio_file_path= 'Audio_Data\\Audio2.wav'
```

```python
#Visualize the wave

import numpy as np
import matplotlib.pyplot as plt
import wave

# Open the WAV file
with wave.open(audio_file_path, 'rb') as wav_file:
    framerate = wav_file.getframerate()
    frames = wav_file.readframes(-1)
    signal = np.frombuffer(frames, dtype=np.int16)

# Create a time axis for the waveform
time = np.linspace(0, len(signal) / framerate, num=len(signal))

# Plot the waveform
plt.figure(figsize=(10, 4))
plt.plot(time, signal)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Waveform of WAV File')
plt.grid(True)
plt.show()
```

```python
# Audio to Vector
import numpy as np
from scipy.io import wavfile


sample_rate, audio_data = wavfile.read(audio_file_path)

# Convert audio data to a numeric vector
audio_vector = audio_data.astype(np.float32)


# Print the length of the audio vector
print("Length of audio vector:", len(audio_vector))
print("sample rate : ", sample_rate)
```

```python
#Play Audio

from IPython.display import Audio

Audio(audio_vector, rate=sample_rate)
```
